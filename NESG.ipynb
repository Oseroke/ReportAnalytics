### Word count

```python
def map_report(filtered_words):
    hash_map={}
    
    if filtered_words is not None:
        for element in filtered_words:
            
            #Remove punctuations
            word = element.replace(',','')
            word = word.replace('.','')
            word = word.replace('(','')
            word = word.replace(')','')
            word = word.replace(':','')
            word = word.replace(';','')
            word = word.replace('%','')
            word = word.replace('Å','')
            word = word.replace('Â»','')
            
            word = word.lower()
            
            #check if word exists
            if word in hash_map:
                hash_map[word] = hash_map[word] + 1
            else:
                hash_map[word] = 1
            
        return hash_map
    else:
        return None
            
```

```python
hash_map = map_report(filtered_words)
hash_map.pop('')
print(len(hash_map))
```

```python
hash_map
```

```python
#sort words by frequency
from operator import itemgetter

sorted_hashmap = sorted(hash_map.items(), key=itemgetter(1), reverse=True)
```
